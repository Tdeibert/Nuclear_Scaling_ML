```python
with tiff.TiffFile(IMAGE) as tif: # open the tiff file for inspection. This creates the object. After the : we are now working inside of the object opened. 

    series = tif.series[0] # we are interogating the tiff file metadata and creating a series object "Series". We can call the object down stream to assign the associated data. 
    axes = series.axes # we are interogating the tiff file metadata and creating a series object "Axes". We can call the object down stream to assign the associated data.
    shape = series.shape # we are interogating the tiff file metadata and creating a series object "Shape". We can call the object down stream to assign the associated data. 
    # we expect the shape to be some form of T,C,Z,X,Y. The shape axes are then associated with the shape dimentions. We can then use the axes to assign the shape dimensions to the correct axis labels. By looking for the shape discriptor associated with the variable we want to assign. For example for time we would look for T in axes. Structure should be axes -> shape -> "Variable" 
    print(f"axes: {axes}, shape: {shape}")

    def get_dim(axes_str, shape_tuple, axis_label): # We are creating a function to assign the shape dimentions to the correct axis labels. This make it dynamic and can used for any image with any shape and axes. we are passing the axes string,shape tuple, and the axis label we want to assign.the fuction will look for the axis label in the axes string and return the corresponding shape dimension. If the axis label is not found, it will return 1. This allows us to handle cases where certain axes may be missing from the image data.
        if axis_label in axes_str:
            inx = axes_str.index(axis_label)
            return shape_tuple[inx]
        return 1 
    
    n_time = get_dim(axes, shape, "T") #call the get_dim function to assign the time points. We are looking for T in the axes string and returning the corresponding shape dimension. If T is not found, it will return 1. This allows us to handle cases where the time axis may be missing from the image data.
    n_z = get_dim(axes, shape, "Z") #call the get_dim function to assign the Z-slices. We are looking for Z in the axes string and returning the corresponding shape dimension. If Z is not found, it will return 1. This allows us to handle cases where the Z-axis may be missing from the image data.
    n_channel = get_dim(axes, shape, "C") #call the get_dim function to assign the channels. We are looking for C in the axes string and returning the corresponding shape dimension. If C is not found, it will return 1. This allows us to handle cases where the channel axis may be missing from the image data.
    print(f"Time points: {n_time}, Z-slices: {n_z}, Channels: {n_channel}") # prints the assigned time points, Z-slices, and channels for the image data. This provides a summary of the dimensions of the image data based on the axes and shape information extracted from the tiff file metadata.
    
```
